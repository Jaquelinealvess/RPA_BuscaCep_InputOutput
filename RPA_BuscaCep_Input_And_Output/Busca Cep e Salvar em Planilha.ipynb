```python
from openpyxl import load_workbook
import os
 
#Pegamos o caminho do arquivo no computador
nome_arquivo_cep = "C:\\Users\jasantos\Curso RPA\RPA_BuscarCep\PesquisaEndereco.xlsx"
planilhaDadosEndereco = load_workbook(nome_arquivo_cep)
 
#Selecionamos a sheet de Dados
sheet_selecionada = planilhaDadosEndereco['CEP']
 
 
 
from selenium import webdriver as opcoesSelenium
from selenium.webdriver.common.keys import Keys
import pyautogui as tempoEspera
 
#Abre o navegador do Google Chrome e abre o site
navegador = opcoesSelenium.Chrome()
navegador.get("https://buscacepinter.correios.com.br/app/endereco/index.php")
 
#Imprime o CEP no campo do CEP
navegador.find_element_by_name("endereco").send_keys("23548057")
 
#Aguarda 2 segundos
tempoEspera.sleep(2)
 
#Clica no botão de Pesquisar
navegador.find_element_by_name("btn_pesquisar").click()
 
#Aguarda 3 segundos
tempoEspera.sleep(3)
 
for i in range(2, len(sheet_selecionada['A']) + 1):
    
    #Aguarda 3 segundos
    tempoEspera.sleep(3)
    
    navegador.find_element_by_id('btn_nbusca').click()
    
    cepPesquisa = sheet_selecionada['A%s' % i].value
    
    #Aguarda 2 segundos
    tempoEspera.sleep(2)
    
    #Imprime o CEP no campo do CEP
    navegador.find_element_by_name("endereco").send_keys(cepPesquisa)
 
    #Aguarda 2 segundos
    tempoEspera.sleep(2)
 
    #Clica no botão de Pesquisar
    navegador.find_element_by_name("btn_pesquisar").click()
 
    #Aguarda 6 segundos
    tempoEspera.sleep(6)
    
    #Pega os dados da Rua no site
    rua = navegador.find_element_by_xpath('//*[@id="resultado-DNEC"]/tbody/tr/td[1]').text
    print("Rua: ", rua)
 
    #Pega os dados do bairro no site
    bairro = navegador.find_element_by_xpath('//*[@id="resultado-DNEC"]/tbody/tr/td[2]').text
    print("Bairro: ", bairro)
 
    #Pega os dados da Cidade no site
    cidade = navegador.find_element_by_xpath('//*[@id="resultado-DNEC"]/tbody/tr/td[3]').text
    print("Cidade: ", cidade)
 
    #Pega os dados do CEP no site
    cep = navegador.find_element_by_xpath('//*[@id="resultado-DNEC"]/tbody/tr/td[4]').text
    print("CEP: ", cep)
    
    #Selecionamos a sheet de Dados
    sheet_Dados = planilhaDadosEndereco['Dados']
 
    #Pegamos a ultima linha preenchida na coluna a e acrescentamos + 1
    linha = len(sheet_Dados['A']) + 1
    colunaA = "A" + str(linha) #Criamos a variavel para juntar A + a ultima linha ex: A2
    colunaB = "B" + str(linha) #Criamos a variavel para juntar B + a ultima linha ex: B2
    colunaC = "C" + str(linha) #Criamos a variavel para juntar C + a ultima linha ex: C2
    colunaD = "D" + str(linha) #Criamos a variavel para juntar D + a ultima linha ex: D2
 
    #Imprimimos as informacoes do site na planilha
    sheet_Dados[colunaA] = rua
    sheet_Dados[colunaB] = bairro
    sheet_Dados[colunaC] = cidade
    sheet_Dados[colunaD] = cep
 
    
#Salvamos o arquivo do Excel com as novas informações
planilhaDadosEndereco.save(filename=nome_arquivo_cep)
 
#Abrimos o arquivo
os.startfile(nome_arquivo_cep)
```
